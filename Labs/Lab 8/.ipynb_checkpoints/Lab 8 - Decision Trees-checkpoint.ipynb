```python
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn â‰¥0.20 is 
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import pandas as pd
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn import datasets
from sklearn.metrics import accuracy_score as accs
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
address = '/Users/sohamkakra/Desktop/F21DL_Portfolio/Datasets/heart.csv'
data = pd.read_csv(address)
labels = data[["output"]]
```

```python
def specificity(y_train, y_train_pred):
    tn, fp, fn, tp = confusion_matrix(y_train, y_train_pred).ravel()
    print("\nSpecificity Score", (tn / (tn+fp)))

def sensitivity(y_train, y_train_pred):
    tn, fp, fn, tp = confusion_matrix(y_train, y_train_pred).ravel()
    print("\nSensitivity Score", (tn / (tn+fn)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['output'], axis=1),
                                                            data['output'], test_size=0.2, random_state = 100)

tree_clasf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clasf.fit(X_train, y_train)

plot_tree(tree_clasf);


y_pred_tree = tree_clasf.predict(X_test)
y_expect = y_test

print("Accuracy = ", accs(y_expect, y_pred_tree))

y_train_pred_tree = cross_val_predict(tree_clasf, X_train, y_train, cv=3)


print("\nPrecision Score = ", precision_score(y_train, y_train_pred_tree))
print("\nRecall Score =", recall_score(y_train, y_train_pred_tree))
print("\nROC Score =", roc_auc_score(y_train, y_train_pred_tree))


specificity(y_train, y_train_pred_tree)
sensitivity(y_train, y_train_pred_tree)
```
